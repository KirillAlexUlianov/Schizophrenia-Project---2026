```r
library("sva")
```

```r
data <- read.csv("data/PsychENCODE-Filtered.csv")
```

```r
individualID <- data$individualID
```

```r
row.names(data) <- data$individualID
```

```r
data['individualID'] <- NULL
```

```r
data_int <- sapply(data, as.integer)
```

```r
meta <- read.csv('data/PsychENCODE-Metadata.csv', row.names = 1)
meta <- meta[row.names(data), ]
```

```r
adjusted <- ComBat_seq(t(data_int), batch=meta[['individualIdSource']])
```

```r
adjusted = t(adjusted)
```

```r
row.names(adjusted) = individualID
```

```r
write.csv(adjusted, file="data/PsychENCODE-ComBat-seq.csv")
```
